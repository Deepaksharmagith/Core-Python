```python
#Class variables can be modified using class methods, and any change will reflect across all instances since the variable is shared
class Student:
    school = "ABC School"   # class variable (shared)

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("School:", Student.school)

    @classmethod
    def change_school(cls, new_school):
        cls.school = new_school   # change class variable


# objects
s1 = Student("Deepak", 22)
s2 = Student("Rahul", 25)

s1.display()
print("------")

# change school for all students
Student.change_school("XYZ School")

s2.display()
```

```python
#If user doesn’t pass school → use class variable (default)
#If user passes school → use user-defined value

class Student:
    school = "ABC School"   # class variable (default)

    def __init__(self, name, age, school=None):
        self.name = name
        self.age = age

        # if user does not pass school → use class variable
        if school is None:
            self.school = Student.school
        else:
            self.school = school

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("School:", self.school)


# objects
s1 = Student("Deepak", 22)                 # no school passed
s2 = Student("Rahul", 25, "XYZ School")    # custom school

s1.display()
print("------")
s2.display()
```

Types of Methods
1. Instance Method-->Works on object (instance) data
def display(self):

2. Class Method-->Works on class data
@classmethod
def change_school(cls):

3. Static Method-->Works like a normal function inside class, No self, cls
@staticmethod
def info():


int--> There are three types of methods in Python: instance methods, class methods, and static methods. Methods defined with @classmethod are called class methods, and they operate on class-level data using the cls parameter

🔥 @staticmethod in Python — What & Why

A static method is a function inside a class that:

does not use self (object)
does not use cls (class)
behaves like a normal utility function, but is logically grouped inside the class

class Student:

    @staticmethod
    def greet():
        print("Welcome to Student System")

Student.greet()

🔥 When to Use @staticmethod

Use it when:
Logic is related to the class
But it doesn’t need object data or class data

Real Example (Data Validation)
class Student:

    @staticmethod
    def is_valid_age(age):
        return age >= 18
print(Student.is_valid_age(20)) 



```python
class Student:
    school = "ABC School"   # class variable

    def __init__(self, name, age, marks):
        self.name = name        # instance variable
        self.age = age
        self.marks = marks

    # 🔹 Instance Method (uses object data)
    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Marks: {self.marks}, School: {Student.school}")

    # 🔹 Class Method (works on class variable)
    @classmethod
    def change_school(cls, new_school):
        cls.school = new_school

    # 🔹 Static Method (utility logic)
    @staticmethod
    def is_pass(marks):
        return marks >= 40


# objects
s1 = Student("Deepak", 22, 85)
s2 = Student("Rahul", 21, 35)

# instance method
s1.display()
s2.display()

print("------")

# static method
print("Deepak Pass:", Student.is_pass(s1.marks))
print("Rahul Pass:", Student.is_pass(s2.marks))

print("------")

# class method
Student.change_school("XYZ School")

s1.display()
s2.display()
```

Without Class Method (Bad Practice)
Works, but:

No validation
No control
Hard to manage in large systems

```python
#apply all 
class Employee:
    bonus_percent = 10   # class variable

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calculate_salary(self):
        bonus = (self.salary * Employee.bonus_percent) / 100
        return self.salary + bonus

    @classmethod
    def update_bonus(cls, new_bonus):
        # ✅ Validation
        if not isinstance(new_bonus, (int, float)):    # isinstance() is used to check the type of an object.
            print("Error: Bonus must be a number")
            return

        if new_bonus < 0 or new_bonus > 100:
            print("Error: Bonus must be between 0 and 100")
            return

        # ✅ Control (single point of update)
        old_bonus = cls.bonus_percent
        cls.bonus_percent = new_bonus

        # ✅ Manageability (logging / tracking)
        print(f"Bonus updated from {old_bonus}% to {new_bonus}%")


# objects
e1 = Employee("Deepak", 50000)
e2 = Employee("Rahul", 60000)

print("Before Update:")
print(e1.calculate_salary())
print(e2.calculate_salary())

print("------")

# update bonus safely
Employee.update_bonus(20)

print("------")

print("After Update:")
print(e1.calculate_salary())
print(e2.calculate_salary())
```
