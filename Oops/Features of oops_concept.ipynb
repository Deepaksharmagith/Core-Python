<span style="color:yellow; font-size:25px;">Abstraction</span>

```python
# hiding unnecesary deatils from users through class, methods.

class Student: # student class
    def __init__(self, name, grade, percentage): 
        self.name = name  # attribute 
        self.grade = grade  
        self.percentage = percentage 

#When we use this, the user cannot see the internal details of the method in the output.
#if you want to give every student an extra 2% grace marks without anyone knowing, you can hide this kind of calculation or logic inside a method.

    def student_details(self): # method - abstraction
        print(f"{self.name} is in class {self.grade}, with {self.percentage+2}%") # hidden from users

# object - instance of class 
student1 = Student('Madhav', 11, 96)
student2 = Student('Vishakha', 12, 97)

#print(student2.percentage)--> actual marks
student2.student_details()
```

<span style="color:yellow; font-size:25px;">Encapsulation</span>

```python
# Restrict access to certain attributes or methods to protect data and enforce controlled access

class Student: 
    def __init__(self, name, grade, percentage): 
        self.name = name   
        self.grade = grade  
        self.__percentage = percentage # double underscore limits access

    def get_percentage(self):    #(use get for best practices)
        return self.__percentage # you can excess private data by creating a saperate method for private data

    def student_details(self): 
        print(f"{self.name} is in class {self.grade}, with {self.percentage}%") 

# object 
student1 = Student('Madhav', 11, 96)
student2 = Student('Vishakha', 12, 97)

#print(student1.__percentage) # error-->percentage attri can't be excess directely
#print(student1.percentage) # error
print(student2.get_percentage())
```

```python
#Suppose you don't want to show % of students whose percentage is less than 70 (% > 70)
class Student: 
    def __init__(self, name, grade, percentage): 
        self.name = name   
        self.grade = grade  
        self.__percentage = percentage 
        
    def get_percentage(self):
        return self.__percentage
        
    def student_details(self):
        if self.__percentage > 70:
            print(f"{self.name} is in class {self.grade}, with {self.__percentage}%") 
        else:
            print(f"{self.name} is in class {self.grade}")

student1 = Student('Madhav', 11, 96)
student2 = Student('Vishakha', 12, 60)


student1.student_details()
student2.student_details()
```

<span style="color:yellow; font-size:25px;">Inheritance</span>

```python
class A:
    def funA(self):
        print("this is a function of a")

class B:        
    def funB(self):
        print("this is a function of b")

class C(A):
    def func(self):
        print("this is function of c")

obj=A()
obj.funA()

obj1=B()
obj1.funB()  

obj2=C()
obj2.func()    
```

```python
# allows one class (child) to reuse the properties and methods of another class (parent). 

# parent class - Father
class Student: 
    def __init__(self, name, grade, percentage): 
        self.name = name  
        self.grade = grade  
        self.percentage = percentage 

    def student_details(self): 
        print(f"{self.name} is in class {self.grade}, with {self.percentage}%")

# object
student1 = Student('Madhav', 11, 96)
student2 = Student('Vishakha', 12, 99)

# child class - beta 
class GraduateStudent(Student): # Graduatestudent child class inherit prop and methods from Student Parent class
    def __init__(self, name, grade, percentage, stream): # old parameters from parent class and new parameters in child class
        super().__init__(name, grade, percentage) # super() is used to access the parent class.
        self.stream = stream # new attribute in child class 

    def student_details(self):
        super().student_details() # method inherit from parent class
        print(f'Stream is {self.stream}')

# object
Grad_Student1 = GraduateStudent('Keshav', 12, 96, 'PCM')
print(Grad_Student1.stream) 

Grad_Student1.student_details()
```

<span style="color:yellow; font-size:25px;">Polymorphism</span>

```python
# allows methods in difft classes to have same name but difft behaviour depending on objects 

class Student: 
    def __init__(self, name, grade, percentage): 
        self.name = name  
        self.grade = grade  
        self.percentage = percentage 

    def student_details(self): 
        print(f"{self.name} is in class {self.grade}, with {self.percentage}%")

# object 
student1 = Student('Madhav', 11, 96)
student2 = Student('Vishakha', 12, 99)

# child class  
class GraduateStudent(Student): 
    def __init__(self, name, grade, percentage, stream): 
        super().__init__(name, grade, percentage) 
        self.stream = stream 

    def student_details(self): 
        print(f'{self.name} is in class {self.grade}, with {self.percentage}% and from stream {self.stream}')
        print('same method with difft o/p')

# object - Student class
student1 = Student('Madhav', 11, 96)

# object - GraduateStudent class
Grad_Student1 = GraduateStudent('Keshav', 12, 96, 'PCM')

student1.student_details()
Grad_Student1.student_details()
```
