1 → If we want to store data for 100 students, we would have to create 100 separate lists, which is not efficient.

2 → Suppose we need to add 15 more details for each student. We would have to use the append() function multiple times, making the program 
    very messy and hard to manage.

3 → Initially, we may store only 2 details (name and grade). Later, if we add 15 more details, it becomes difficult to remember which 
    data is stored at which position while accessing it.

4 → Sequence problem: Different people may store data in different orders (e.g., some may store the student’s name first, 
    others may store the class first), leading to inconsistency.

5 → Structure problem: If we try to add teacher details along with student details, the structure becomes confusing and unorganized.

6 → Python does not know that student1 and student2 are related. It treats them as separate variables. If a teacher mistakenly adds 
    only one detail ( just student’s name), then we cannot perform operations properly because that variable is a string instead of a list.

7 → Data privacy cannot be maintained properly using this approach.

8 → This structure is not reusable for another classroom or similar use case.

```python
# using OOPs- creating student records

# __init__ method - constructor, init(initilise) is a fixte method in python which inisalize the obj value/attributes.
# self parameter - reference or connection build btw class and object

class Student: 
    def __init__(self, name, grade, percentage):  # method
        self.name = name  # attribute 
        self.grade = grade # attribute 
        self.percentage = percentage # attribute


# object 
student1 = Student('Madhav', 11, 96)
print(student1.name, student1.grade)

student2 = Student('Vishakha', 12, 99)
print(student2.name, student2.grade)
```

```python
class Student: 
    def __init__(self, name, grade):  
        self.name = name  
        self.grade = grade 

    def student_details(self): # method 
        print(f"{self.name} is in class {self.grade}")

# object 
student1 = Student('Madhav', 11)
student2 = Student('Vishakha', 12)

#call method through obj--> student1
#can we call a method  outside the class through object?--> Yes--> mostly hum wahi karte hain 
student1.student_details()
student2.student_details()
```

```python
class Student: 
    def __init__(self, name, grade, percentage):  
        self.name = name  
        self.grade = grade 
        self.percentage = percentage 

    def student_details(self): # method 
        print(f"{self.name} is in class {self.grade}, with {self.percentage}%")

# object 
student1 = Student('Madhav', 11, 96)
student2 = Student('Vishakha', 12, 99)
```

```python
# call a method to execute it (obj_name.method_name)
student1.student_details()
student2.student_details()
```

<span style="color:yellow; font-size:25px;">Creating variables (team1, team2) and establishing a relationship (link) with an object.</span>

```python
class Studentt: 
    def __init__(self, name, grade, percentage, team):  
        self.name = name   
        self.grade = grade  
        self.percentage = percentage 
        self.team = team

    def student_details(self):  
        print(f"{self.name} is in class {self.grade}, with {self.percentage}% and is in team {self.team}")

team1 = 'A'
team2 = 'B'
# Now we assign these teams to students. So, we will add a new attribute: team.

student1 = Studentt('Madhav', 11, 96, team1)
student2 = Studentt('Vishakha', 12, 99, team2)


print(student2.team)
```

```python
student1.student_details()
student2.student_details()
```
