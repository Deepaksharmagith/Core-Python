<span style="color:yellow; font-size:25px;">Break Statement</span>
--> we appies condition for break

```python
# print 10 to 25 no with break
n=int(input("enter a no"))                           
i=10                                                 
while(i<=25):                                          
    if(i==n):                                              
        break                                                        
    print(i)                                                
    i=i+1   
#i=i+1           #if you write inciment out side the loop, it will give 1 infinit times bez i ka incriment hua hi nahi 
```

```python
n=int(input("enter a no"))                      # 11 kyu aya bez jb i=10 pass hoga to 11 print hoga(i=10+1)
i=1                                            
while(i<=10):
    i=i+1
    if(i==n):                                           
        continue                                                       
    print(i)        

    # use list also
```

```python
# Escape Multiple NO, using user input
n=int(input("enter a no"))  
n2=int(input("enter a no"))
i=1                                            
while(i<=10):
    i=i+1
    if(i==n):                                           
        continue 
    if(i==n2):
        continue
    print(i) 
```

<span style="color:yellow; font-size:25px;">For LOOP</span>

```python
#(1) print no from 1 to 5
for i in range(1, 6): #auto incriment
    print(i)
```

```python
#(2) print every 3 no b/w 1 to 100
for i in range (1,100,3):
    print(i, end= " ")
```

```python
#(3) print reverse no
for i in range(10, 0, -1):
    print(i, end = " ")
```

```python
#(4) apply loop on string
for i in 'deepak':
    print(i)
# deepak is a string i str ki har ek value pr chalaga
```
