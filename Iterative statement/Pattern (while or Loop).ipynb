```python
i=1
while(i<=5):   #<---count rows
#---------------------------------------
    j=1
    while(j<=5-i):  #<---blank space
        print(' ',end='  ')
        j=j+1
#---------------------------------------
    k=1
    while(k<=i):   #<---print star
        print('*',end='  ')
        k=k+1
#---------------------------------        
    print()
    i=i+1
```

```python
i=1
k=1
while(i<=5):  
    j=1
    while(j<=5-i):  
        print('_',end='  ')
        j=j+1
    b=1
    while(b<=k):  
        print('*',end='  ')
        b=b+1
    k=k+2  
    print()
    i=i+1
```
