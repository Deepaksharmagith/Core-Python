```python
# 1. Find sum of digits of a given number.

n=int(input("enter digits")) 
sum=0
while(n>0):
    sum=sum+n%10
    n=n//10
print("Sum of digits",sum)
    
# n=int(input("enter digits")) 
# i=1
# sum=0
# while(i<=n):
#     sum=sum+n%10
#     n=n//10
#     print(sum)
```

```python
# 2. Find product of digits of a given number.
n=int(input("enter digits")) 
prod=1
while(n>0):
    prod=prod*n%10
    n=n//10
print("Product of digits",prod)
```

```python
# 3. Find sum of even digits and product of odd digits of a given number.  (same for numbers 1- n)
n=int(input("enter digits")) 
sum=0
prod=1
while(n>0):
    if n%2==0:
        sum=sum+n%10
        n=n//10
    else:
        prod=prod*n%10
        n=n//10
print("sum of even digits",sum)
print("Product of odd digits",prod)


# while(n>0):
#     d=i%10
#     if n%2==0:
#         sum=sum+d
#     else:
#         prod=prod*d
#     n=n//10

```

```python
# 4. Find sum of square of digits of a given number.
n=int(input("enter digits")) 
sum=0
while(n>0):
    sum=sum+(n%10)*(n%10)
    n=n//10
print("sum of square of digits",sum)
```

```python
# 5. Find sum of cube of digits of a given number.
n=int(input("enter digits")) 
sum=0
while(n>0):
    sum=sum+(n%10)*(n%10)*(n%10)
    n=n//10
print("sum of cube of digits",sum)
```

```python
# 6. Check whether a given number is armstrong or not.
n=int(input("enter digits")) 
org =n
org1=n
count=0
sum=0
while(n>0):
    n=n//10
    count=count+1
print(count)


while(org>0):
    sum=sum+(org%10)*(org%10)*(org%10)
    org=org//10
print(sum)

if org1==sum:
    print("NO is Arm")
else:
    print("No is Not")
    


```

```python
# 7. Reverse a given number.
n=int(input("enter digits")) 
Rev=0
while(n>0):
    Rev=(Rev*10)+n%10
    n=n//10
print(Rev)
```

```python
# 8. Check whether a given number is palindrome number or not.
n=int(input("enter digits")) 
org=n
Rev=0
while(n>0):
    Rev=(Rev*10)+n%10
    n=n//10
print(Rev)
if org==Rev:
    print("NO is pali")
else:
    print("No is Not")
```

```python
# 9. Print factorial of given number. 5!-->5x4x3x2x1=120
n=int(input("enter a no"))
fac=1
while(n>0):
    fac=fac*n
    n=n-1
print("factorial is", fac)
```
