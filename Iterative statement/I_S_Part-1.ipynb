```python
#(3) Check whether a no is prime or composit, NO given by user (using while & for)
#While
n=int(input("enter a no"))
prime=True
i=2
while(i<n):
    if(n%i==0):
        prime=False       
        break               # remove break , it will stil work
    i=i+1
if(prime==True):
    print("no is prime")
else:
    print("no is composit")
```

```python
#(4) print all prime no b/w 1 to 100
for i in range(1, 100):
    prime= True
    for k in range(2,i):
        if(i%k==0):
            prime = False
            break
    if(prime):
        print(i, end=" ")
```

```python
#(5) Searching
l=[]
size=int(input("enter the size of list"))
for i in range(0,size):
    l.append(int(input("enter the element")))
print(l)
```

```python
search=int(input("enter the search no"))
found=False
for k in l:
    if(k==search):
        break
if(found):
    print("search element found")
else:
    print("search element not found")
```
