```python
# 5. Print sum from 1 to n

n=int(input("Enter a no"))
i=1
sum=0
while(i<=n):
    sum=sum+i
    i=i+1
print("Sum from 1 to", n, "is:", sum)
```

```python
# 6. Print sum from n to 1

n = int(input("Enter a no: "))
sum = 0

while n >= 1:
    sum = sum + n
    n = n - 1
    print(sum)
```

```python
# 7. Print sum of squares from 1 to n  

n=int(input("Enter a no"))
i=1
sum=0
while(i<=n):
    sum=sum+i*i
    i=i+1
print("Sum of squares",sum)

print()
# 8. Print sum of cubes from 1 to n

n=int(input("Enter a no"))
i=1
sum=0
while(i<=n):
    sum=sum+i*i*i
    i=i+1
print("Sum of cubes",sum)
```

```python
# 10. Print all numbers b/w 200 to 300 ending with 3 or 7

i = 200
while(i <= 300):
    if(i % 10 == 3) or (i % 10 == 7):
        print(i, end=", ")
    i = i + 1

# OR
# i=200
# while(i<=300):
#     if(i%10==3):
#         print(i, end=",")
#     if(i%10==7):
#         print(i, end=",")
#     i=i+1

# OR    
print()

for n in range(200, 301):
    if n % 10 == 3 or n % 10 == 7:
        print(n, end=", ")
```

```python
# 11. Find sum of even numbers from 1 to n

n = int(input("Enter a no: "))
i = 1
sum = 0

while(i <= n):
    if(i % 2 == 0):
        sum = sum + i
    i = i + 1

print("Sum of even numbers:", sum)
        
```
