```python
#1 Check whether a no is prime or composit, 

#prime no / by 1 or it self meant a prime no has 2 factors only

#While
n=int(input("enter a no"))
prime=True
i=2
while(i<n):
    if(n%i==0):
        prime=False       
        break               # remove break , it will stil work
    i=i+1
if(prime==True):
    print("no is prime")
else:
    print("no is composit")

#For
n=int(input("enter a no"))
prime=True
for i in range(2,n):
    if(n%i==0):
        prime=False
        break                                       # doubt use of break
if(prime==True):
    print("no is prime")
else:
    print("no is composit")
```
