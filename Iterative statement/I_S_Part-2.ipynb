```python
#(6) Print sum of even and product of odd digits
n=int(input("enter a no"))
s=0
product=1
while(n>0):
    d=n%10
    if(d%2==0):
        s=s+d
    else:
        product=product*d
    n=n//10
print(s,product)  
```

```python
#(7) print sum of cube from 1 to n or sum of cube of digits of a given no
n=int(input("enter a no"))
sum=0
while(n>0):
    sum=sum+(n%10)*(n%10)*(n%10)
    n=n//10
    print(sum)
```

```python
#(8) print sum of square or sum of square of digits of a given no
n=int(input("enter a no"))
sum=0
i=1
while(i<=n):
    sum=sum+i*i
    i=i+1
print(sum)
```
